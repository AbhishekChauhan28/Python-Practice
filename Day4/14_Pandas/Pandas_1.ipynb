# How to select a Column from a DataFrame?

```python
# read a dataset of UFO reports into a DataFrame
ufo = pd.read_csv('data/ufo.csv')
```

```python
# examine the first 5 rows
ufo.head()
```

```python
# select the 'City' Column using bracket notation
# Column also called as Series
ufo['City']
```

```python
# or equivalently, use dot notation
ufo.City
```

**Square Bracket notation** will always work, whereas **dot notation** has limitations:

- Dot notation doesn't work if there are **spaces** in the Column name
- Dot notation doesn't work if the Column has the same name as a **DataFrame method or attribute** (like 'head' or 'shape')
- Dot notation can't be used to define the name of a **new Column**

```python
# create a new 'Location' Column (must use bracket notation to define the Column name)
ufo['Location'] = ufo.City + ', ' + ufo.State
ufo.head()
```

```python
ufo[5:12]
```

# How to filter rows of a pandas DataFrame by column value?

```python
movies.head()
```

```python
# examine the number of rows and columns
movies.shape
```

**Goal:** Filter the DataFrame rows to only show movies with a 'duration' of at least 200 minutes.

```python
type(movies.duration)
```

```python
# create a list in which each element refers to a DataFrame row: True if the row satisfies the condition, False otherwise
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
# confirm that the list has the same length as the DataFrame
len(booleans)
```

```python
booleans[:10]
```

```python
# use bracket notation with the boolean Series to tell the DataFrame which rows to display
movies[booleans]
```

```python
# simplify the steps above: no need to write a for loop to create 'is_long' 
# pandas will broadcast the comparison
is_long = movies.duration >= 200
movies[is_long]
```

```python
# or equivalently, write it in one line (no need to create the 'is_long' object)
movies[movies.duration >= 200]
```

```python
# select the 'genre' Series from the filtered DataFrame
movies[movies.duration >= 200].genre
```

```python
movies["Len_title"] = movies.title.str.len()
```

```python
movies.sort_values(["Len_title","title"]).head(20)
```

# How to apply multiple filter criteria to a pandas DataFrame?

```python
movies.head()
```

```python
# filter the DataFrame to only show movies with a 'duration' of at least 200 minutes
movies[movies.duration >= 200]
```

**logical operators:**

- **`and`**: True only if **both sides** of the operator are True
- **`or`**: True if **either side** of the operator is True

Rules for specifying **multiple filter criteria** in pandas:

- use **`&`** instead of **`and`**
- use **`|`** instead of **`or`**
- add **parentheses** around each condition to specify evaluation order

Q. Filter the DataFrame of long movies (duration >= 200) to only show movies which also have a 'genre' of 'Drama'

```python
# use the '&' operator to specify that both conditions are required
movies[(movies.duration >=200) & (movies.genre == 'Drama')]
```

Q. Filter the original DataFrame to show movies with a 'genre' of 'Crime' or 'Drama' or 'Action'

```python
# use the '|' operator to specify that a row can match any of the three criteria
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')].head(10)
```

```python
# or equivalently, use the 'isin' method
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])].head(10)
```
